## Step 4 - Categorize Each Country

Create a new column "Group". If a country is one of the seven G7 countried, the value of "Group" would be "G7", if a country of one of the five BRICS countries, the value of "Group" would be "BRICS". Otherwise, the value of "Group" would be "REST of WORLD".

```python
# Lists of G7 and BRICS countries
g7_countries = ['Canada', 'France', 'Germany', 'Italy', 'Japan', 'United Kingdom', 'United States']
brics_countries = ['Brazil', 'Russian Federation', 'India', 'China', 'South Africa']

```

```python
# Categorize each country based on the group
def categorize_country(country):
    if country in g7_countries:
        return "G7"
    elif country in brics_countries:
        return "BRICS"
    else:
        return "REST of WORLD"

# Apply the function to create the new column
df['Group'] = df['Country Name'].apply(categorize_country)

# Display sample rows to confirm
print(df[['Country Name', 'Group']].sample(10))
```

## Step 5 - Aggregate Data

Aggregate the following columns at the "Group" level for each year:

- For GDP, using "sum" for aggregation.
- For GDP Per Capita, use "mean" for aggregation.
- Check your calculation with the reference site.

```python
# Aggregate GDP (sum) and GDP Per Capita (mean) at the Group level for each year
aggregated_data = df.groupby(['Year', 'Group']).agg({
    'GDP ($Billions)': 'sum',
    'GDP Per Capita ($)': 'mean'
}).reset_index()
aggregated_data.rename(columns={'GDP ($Billions)': 'Total GDP', 'GDP Per Capita ($)': 'Average GDP Per Capita'}, inplace=True)

# Display the aggregated data
print(aggregated_data)
```

## Step 6 - Calculate Percentage Share of GDP

For each year, calculate the percentage share of each group's GDP. Check your calculation with the reference site.

```python
# Calculate the total GDP across all groups for each year
total_gdp_per_year = aggregated_data.groupby('Year')['Total GDP'].sum().reset_index()
total_gdp_per_year.rename(columns={'Total GDP': 'Yearly Total GDP'}, inplace=True)

# Merge the yearly total GDP with the aggregated data
aggregated_data = pd.merge(aggregated_data, total_gdp_per_year, on='Year')

# Calculate the percentage share for each group
aggregated_data['GDP Percentage Share'] = (aggregated_data['Total GDP'] / aggregated_data['Yearly Total GDP']) * 100

# Display the data
print(aggregated_data)
```

## Step 7 - Plot the GDP Percentage Share Over Time

```python
import plotly.express as px

# Create a line plot for GDP Percentage Share over time for each group
fig = px.line(
    aggregated_data,
    x='Year',
    y='GDP Percentage Share',
    color='Group',
    title='GDP Percentage Share Over Time by Group',
    labels={'GDP Percentage Share': '% Share of GDP', 'Year': 'Year'},
    markers=True  # Adds markers to the line for clarity
)

# Show the plot
fig.show()

```

```python
fig.update_traces(line=dict(width=2), marker=dict(size=6))
fig.update_layout(
    xaxis_title="Year",
    yaxis_title="% Share of GDP",
    title_font=dict(size=20),
    legend_title="Group",
    legend=dict(font=dict(size=14))
)

fig.show()
```

## Step 10 - Plot the 2023 GDP Pie Charts

One pie chart for G7 and one pie chart for BRICK. Each country is a slice of a pie. Make sure the following information is displayed for each slice:

- Country Name
- GDP Percentage Share
- GDP Total

```python
# Import necessary libraries
import pandas as pd
import plotly.express as px

url = "https://raw.githubusercontent.com/SaichaitanyavarmaIndukuri-github/UMBC-DATA609-FALL24-WANG/refs/heads/main/2024-11-07T00-25_export.csv"
df = pd.read_csv(url)

# Filter data for 2023
df_2023 = df[df['Year'] == 2023].dropna(subset=['GDP ($Billions)', 'Country Name'])

# Define G7 and BRICS countries
g7_countries = ['Canada', 'France', 'Germany', 'Italy', 'Japan', 'United Kingdom', 'United States']
brics_countries = ['Brazil', 'Russian Federation', 'India', 'China', 'South Africa']

# Create a 'Group' column for categorizing countries
df_2023['Group'] = df_2023['Country Name'].apply(
    lambda x: 'G7' if x in g7_countries else ('BRICS' if x in brics_countries else 'REST of WORLD')
)

# Filter data for G7 and BRICS countries
g7_countries_2023 = df_2023[df_2023['Group'] == 'G7']
brics_countries_2023 = df_2023[df_2023['Group'] == 'BRICS']

# Calculate the total GDP for the year 2023 for each group
total_gdp_g7 = g7_countries_2023['GDP ($Billions)'].sum()
total_gdp_brics = brics_countries_2023['GDP ($Billions)'].sum()

# Calculate GDP percentage share for each country
g7_countries_2023['GDP Percentage Share'] = g7_countries_2023['GDP ($Billions)'] / total_gdp_g7 * 100
brics_countries_2023['GDP Percentage Share'] = brics_countries_2023['GDP ($Billions)'] / total_gdp_brics * 100

# Step 1: Plot the GDP Pie Chart for G7 countries
fig_g7 = px.pie(
    g7_countries_2023,
    names='Country Name',  # Country for each slice
    values='GDP ($Billions)',  # GDP size for each slice
    hover_data={
        'GDP ($Billions)': True,  # Display GDP value on hover
        'GDP Percentage Share': True  # Display GDP percentage share on hover
    },
    title='G7 Countries GDP Share in 2023'
)



# Show the G7 Pie Chart
fig_g7.show()

# Step 2: Plot the GDP Pie Chart for BRICS countries
fig_brics = px.pie(
    brics_countries_2023,
    names='Country Name',  # Country for each slice
    values='GDP ($Billions)',  # GDP size for each slice
    hover_data={
        'GDP ($Billions)': True,  # Display GDP value on hover
        'GDP Percentage Share': True  # Display GDP percentage share on hover
    },
    title='BRICS Countries GDP Share in 2023'
)


# Show the BRICS Pie Chart
fig_brics.show()
```

## Step 15 - Plot the 2023 GDP Per Capita Bar Chart of G7 and BRICS countries

Each country is a bar. Differentiate group by color.

```python
import pandas as pd
import plotly.express as px

df = pd.read_csv("https://raw.githubusercontent.com/SaichaitanyavarmaIndukuri-github/UMBC-DATA609-FALL24-WANG/refs/heads/main/2024-11-07T00-25_export.csv")

# Create a 'Group' column to classify countries into G7, BRICS, or Rest of World
g7_countries = ['Canada', 'France', 'Germany', 'Italy', 'Japan', 'United Kingdom', 'United States']
brics_countries = ['Brazil', 'Russia', 'India', 'China', 'South Africa']

df['Group'] = df['Country Name'].apply(lambda x: 'G7' if x in g7_countries else ('BRICS' if x in brics_countries else 'Rest of World'))

# Filter for 2023 data
df_2023 = df[df['Year'] == 2023]

# Aggregate the data by 'Group' and 'Country Name'
df_aggregated = df_2023.groupby(['Group', 'Country Name'], as_index=False).agg({
    'GDP ($Billions)': 'sum',
    'GDP Per Capita ($)': 'mean'
})

# Filter for G7 and BRICS countries only
df_2023_g7_brics = df_aggregated[df_aggregated['Group'].isin(['G7', 'BRICS'])]

# Create a bar chart for GDP Per Capita in 2023 for G7 and BRICS countries
fig_2023_gdp_per_capita_g7_brics = px.bar(
    df_2023_g7_brics,
    x='Country Name',  # Country on x-axis
    y='GDP Per Capita ($)',  # GDP per Capita on y-axis
    color='Group',  # Color by G7 or BRICS
    title="2023 GDP Per Capita by Country (G7 & BRICS)",
    labels={"GDP Per Capita ($)": "GDP Per Capita ($)", "Country Name": "Country", "Group": "Group"},
    height=600
)

# Show the plot
fig_2023_gdp_per_capita_g7_brics.show()

```
