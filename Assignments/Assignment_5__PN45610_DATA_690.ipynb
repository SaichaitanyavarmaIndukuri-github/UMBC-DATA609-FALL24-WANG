## HW Question 2

Find out if there is a relationship between wealth and health

Pick any year worth of data and use plotly express to generate a scatter plot with GDP as x and life expectancy as y.

```python
import pandas as pd
import plotly.express as px

df = pd.read_csv("https://raw.githubusercontent.com/wcj365/python-stats-dataviz/refs/heads/master/fall2024/data/World_Development_Indicators_(WDI).csv")

year = 2015

# Filter data for the selected year
df_year = df[df['Year'] == year]

df_year = df_year[['GDP per capita (current US$)', 'Life expectancy at birth, total (years)']].dropna()

# Create a scatter plot with Plotly Express
fig = px.scatter(df_year, x='GDP per capita (current US$)', y='Life expectancy at birth, total (years)',
                 title=f'Relationship between GDP and Life Expectancy in {year}',
                 labels={'GDP per capita (current US$)': 'GDP (Current US$)', 'Life expectancy at birth, total (years)': 'Life Expectancy (Years)'})
fig.show()
```

##### We can observe people with more than 50K GDP has life expectancy more than 75 years for sure , and we can observe positive corelation here between GDP and Life expectancy.

## Homework Question 3

Select a year of data. calculate total population of each region and make a pie chart to display the distribution of total population of all regions.

```python
import pandas as pd
import plotly.express as px

df = pd.read_csv("https://raw.githubusercontent.com/wcj365/python-stats-dataviz/refs/heads/master/fall2024/data/World_Development_Indicators_(WDI).csv")
year = 2015

df_year = df[df['Year'] == year]

df_population_by_region = df_year.groupby('Region')['Population, total'].sum().reset_index()

# Create a pie chart with Plotly Express
fig = px.pie(df_population_by_region, values='Population, total', names='Region',
             title=f'Population Distribution by Region in {year}')

fig.show()

```

## Home work question 4

1. remove the color legend.
2. Order the above bar chart based on the values of the average GDP per capita.

```python

import pandas as pd

import plotly.express as px

df = pd.read_csv("https://raw.githubusercontent.com/wcj365/python-stats-dataviz/refs/heads/master/fall2024/data/World_Development_Indicators_(WDI).csv")

year = 2015

df_year = df[df['Year'] == year]


df_year['GDP per Capita'] = df_year['GDP per capita (current US$)'] / df_year['Population, total']

df_avg_gdp_per_capita = df_year.groupby('Region')['GDP per Capita'].mean().reset_index()

df_avg_gdp_per_capita = df_avg_gdp_per_capita.sort_values(by='GDP per Capita', ascending=False)

# Create a bar chart without a color legend and sorted by GDP per Capita
fig = px.bar(df_avg_gdp_per_capita, x='Region', y='GDP per Capita',
             title=f'Average GDP per Capita by Region in {year}',
             labels={'Region': 'Region', 'GDP per Capita': 'Average GDP per Capita (US$)'})

# Remove the color legend
fig.update_traces(marker_color='black', showlegend=False)

# Order the categories based on GDP per Capita values
fig.update_layout(xaxis={'categoryorder':'total descending'})

fig.show()
```

#### 1. Create a line chart showing the trend of Life Expectancy over time for comparing those countries.

```python
import pandas as pd
import plotly.express as px

df = pd.read_csv("https://raw.githubusercontent.com/wcj365/python-stats-dataviz/refs/heads/master/fall2024/data/World_Development_Indicators_(WDI).csv")

countries_of_interest = ['Denmark', 'Finland', 'Iceland', 'Norway', 'Sweden']

df_filtered = df[df['Country'].isin(countries_of_interest)][['Country', 'Year', 'Life expectancy at birth, total (years)']].dropna()

# Create a line chart with Plotly Express
fig = px.line(df_filtered, x='Year', y='Life expectancy at birth, total (years)', color='Country',
              title='Trend of Life Expectancy Over Time for Nordic Countries',
              labels={'Life expectancy at birth, total (years)': 'Life Expectancy (Years)', 'Year': 'Year'})

fig.show()
```

#### 2. Pick a year from the data of the above few countries and create a scatter plot to compare their GDP per capita and life expectancy. The size of the dots represent the population. This type of scatter plot is called bubble chart.

```python
import pandas as pd
import plotly.express as px

df = pd.read_csv("https://raw.githubusercontent.com/wcj365/python-stats-dataviz/refs/heads/master/fall2024/data/World_Development_Indicators_(WDI).csv")

nordic_countries = ['Denmark', 'Finland', 'Iceland', 'Norway', 'Sweden']

year = 2017

df_filtered = df[(df['Country'].isin(nordic_countries)) & (df['Year'] == year)][['Country', 'GDP per capita (current US$)', 'Life expectancy at birth, total (years)', 'Population, total']].dropna()

# Create a bubble chart with Plotly Express
fig = px.scatter(df_filtered,
                 x='GDP per capita (current US$)',
                 y='Life expectancy at birth, total (years)',
                 size='Population, total',
                 color='Country',
                 title=f'GDP per capita (current US$) vs Life expectancy at birth, total (years)({year}) for Nordic Countries',
                 labels={'GDP per capita (current US$)': 'GDP per Capita (US$)', 'Life expectancy at birth, total (years)': 'Life Expectancy (Years)'},
                 hover_name='Country',
                 size_max=69)
fig.show()
```

#### 3. Bonus - compute the average GDP per capita, the average life expectancy, and the total population for each year of all the seven regions (aggregation) and repeat 1 and 2 using the aggregated data for all regions

```python
import pandas as pd
import plotly.express as px

# Load the dataset
df = pd.read_csv("https://raw.githubusercontent.com/wcj365/python-stats-dataviz/refs/heads/master/fall2024/data/World_Development_Indicators_(WDI).csv")

# List of the seven regions
regions = ['East Asia & Pacific', 'Europe & Central Asia', 'Latin America & Caribbean',
           'Middle East & North Africa', 'North America', 'South Asia', 'Sub-Saharan Africa']

df_aggregated = df.groupby(['Year', 'Region']).agg(
    avg_gdp_per_capita=('GDP per capita (current US$)', 'mean'),
    avg_life_expectancy=('Life expectancy at birth, total (years)', 'mean'),
    total_population=('Population, total', 'sum')
).reset_index()

# Step 1: Line chart showing trend of life expectancy over time for the seven regions
fig1 = px.line(df_aggregated,
               x='Year',
               y='avg_life_expectancy',
               color='Region',
               title='Trend of Life Expectancy Over Time for Aggregated Regions',
               labels={'avg_life_expectancy': 'Average Life Expectancy (Years)', 'Year': 'Year'})

fig1.show()

# Step 2: Bubble chart comparing GDP per capita and life expectancy for a specific year
year = 2017

# Filter data for the selected year
df_year_aggregated = df_aggregated[df_aggregated['Year'] == year]

fig2 = px.scatter(df_year_aggregated,
                  x='avg_gdp_per_capita',
                  y='avg_life_expectancy',
                  size='total_population',
                  color='Region',
                  title=f'GDP per capita (current US$) vs Life expectancy at birth, total (years) ({year}) for Aggregated Regions',
                  labels={'avg_gdp_per_capita': 'Average GDP per Capita (US$)', 'avg_life_expectancy': 'Average Life Expectancy (Years)'},
                  hover_name='Region',
                  size_max=69)

fig2.show()

```
