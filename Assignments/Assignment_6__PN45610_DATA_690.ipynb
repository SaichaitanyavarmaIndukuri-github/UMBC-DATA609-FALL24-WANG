## HW Question 2

Find out if there is a relationship between wealth and health

Pick any year worth of data and use plotly express to generate a scatter plot with GDP as x and life expectancy as y.

```python
import pandas as pd
import plotly.express as px

df = pd.read_csv("https://raw.githubusercontent.com/wcj365/python-stats-dataviz/refs/heads/master/fall2024/data/World_Development_Indicators_(WDI).csv")

year = 2015

# Filter data for the selected year
df_year = df[df['Year'] == year]

df_year = df_year[['GDP per capita (current US$)', 'Life expectancy at birth, total (years)']].dropna()

# Create a scatter plot with Plotly Express
fig = px.scatter(df_year, x='GDP per capita (current US$)', y='Life expectancy at birth, total (years)',
                 title=f'Relationship between GDP and Life Expectancy in {year}',
                 labels={'GDP per capita (current US$)': 'GDP (Current US$)', 'Life expectancy at birth, total (years)': 'Life Expectancy (Years)'})
fig.show()
```

## Homework Question 3

Select a year of data. calculate total population of each region and make a pie chart to display the distribution of total population of all regions.

```python
import pandas as pd
import plotly.express as px

df = pd.read_csv("https://raw.githubusercontent.com/wcj365/python-stats-dataviz/refs/heads/master/fall2024/data/World_Development_Indicators_(WDI).csv")
year = 2015

df_year = df[df['Year'] == year]

df_population_by_region = df_year.groupby('Region')['Population, total'].sum().reset_index()

# Create a pie chart with Plotly Express
fig = px.pie(df_population_by_region, values='Population, total', names='Region',
             title=f'Population Distribution by Region in {year}')

fig.show()

```

#### 2. Pick a year from the data of the above few countries and create a scatter plot to compare their GDP per capita and life expectancy. The size of the dots represent the population. This type of scatter plot is called bubble chart.

```python
import pandas as pd
import plotly.express as px

df = pd.read_csv("https://raw.githubusercontent.com/wcj365/python-stats-dataviz/refs/heads/master/fall2024/data/World_Development_Indicators_(WDI).csv")

nordic_countries = ['Denmark', 'Finland', 'Iceland', 'Norway', 'Sweden']

year = 2017

df_filtered = df[(df['Country'].isin(nordic_countries)) & (df['Year'] == year)][['Country', 'GDP per capita (current US$)', 'Life expectancy at birth, total (years)', 'Population, total']].dropna()

# Create a bubble chart with Plotly Express
fig = px.scatter(df_filtered,
                 x='GDP per capita (current US$)',
                 y='Life expectancy at birth, total (years)',
                 size='Population, total',
                 color='Country',
                 title=f'GDP per capita (current US$) vs Life expectancy at birth, total (years)({year}) for Nordic Countries',
                 labels={'GDP per capita (current US$)': 'GDP per Capita (US$)', 'Life expectancy at birth, total (years)': 'Life Expectancy (Years)'},
                 hover_name='Country',
                 size_max=69)
fig.show()
```

#### 3. Bonus - compute the average GDP per capita, the average life expectancy, and the total population for each year of all the seven regions (aggregation) and repeat 1 and 2 using the aggregated data for all regions

```python
import pandas as pd
import plotly.express as px

# Load the dataset
df = pd.read_csv("https://raw.githubusercontent.com/wcj365/python-stats-dataviz/refs/heads/master/fall2024/data/World_Development_Indicators_(WDI).csv")

# List of the seven regions
regions = ['East Asia & Pacific', 'Europe & Central Asia', 'Latin America & Caribbean',
           'Middle East & North Africa', 'North America', 'South Asia', 'Sub-Saharan Africa']

df_aggregated = df.groupby(['Year', 'Region']).agg(
    avg_gdp_per_capita=('GDP per capita (current US$)', 'mean'),
    avg_life_expectancy=('Life expectancy at birth, total (years)', 'mean'),
    total_population=('Population, total', 'sum')
).reset_index()

# Step 1: Line chart showing trend of life expectancy over time for the seven regions
fig1 = px.line(df_aggregated,
               x='Year',
               y='avg_life_expectancy',
               color='Region',
               title='Trend of Life Expectancy Over Time for Aggregated Regions',
               labels={'avg_life_expectancy': 'Average Life Expectancy (Years)', 'Year': 'Year'})

fig1.show()

# Step 2: Bubble chart comparing GDP per capita and life expectancy for a specific year
year = 2017

# Filter data for the selected year
df_year_aggregated = df_aggregated[df_aggregated['Year'] == year]

fig2 = px.scatter(df_year_aggregated,
                  x='avg_gdp_per_capita',
                  y='avg_life_expectancy',
                  size='total_population',
                  color='Region',
                  title=f'GDP per capita (current US$) vs Life expectancy at birth, total (years) ({year}) for Aggregated Regions',
                  labels={'avg_gdp_per_capita': 'Average GDP per Capita (US$)', 'avg_life_expectancy': 'Average Life Expectancy (Years)'},
                  hover_name='Region',
                  size_max=69)

fig2.show()

```

#### Question 1

for the Quad time series comparison chart above, try to add country name to the end of the line and then remove the color legend.

```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go

df = pd.read_csv("https://raw.githubusercontent.com/wcj365/python-stats-dataviz/refs/heads/master/fall2024/data/World_Development_Indicators_(WDI).csv")

# Filter data for the four Quad countries (USA, CHN, IND, RUS)
df_quad = df[df["Country Code"].isin(["USA", "CHN", "IND", "RUS"])]

# Create a line chart for GDP per Capita over time for Quad countries
fig = px.line(df_quad,
              x='Year',
              y='GDP per capita (current US$)',
              color='Country',
              labels={
                  'GDP per capita (current US$)': 'GDP per Capita (Current US$)',
                  'Year': 'Year'
              },
              title='GDP per Capita Over Time for Quad Countries')

# Remove the color legend
fig.update_layout(showlegend=False)

# Add country labels to the end of each line
quad_countries = ["USA", "CHN", "IND", "RUS"]
for country in quad_countries:
    # Get the last data point for each country (most recent year)
    country_data = df_quad[df_quad["Country Code"] == country]
    last_year = country_data["Year"].max()
    last_value = country_data[country_data["Year"] == last_year]["GDP per capita (current US$)"].values[0]

    # Add annotation (text label) at the end of each line
    fig.add_annotation(
        x=last_year,
        y=last_value,
        text=country,
        showarrow=False,
        xanchor='left',
        yanchor='middle',
        font=dict(size=12, color='black')
    )
fig.show()
```

### Question 2

Round the Average GDP for each region to 2 decimal points. Sort the bar according to GDP either in ascending order. Currently, it is ordered alphabetically.

```python
import pandas as pd
import plotly.express as px

# Load the dataset
df = pd.read_csv("https://raw.githubusercontent.com/wcj365/python-stats-dataviz/refs/heads/master/fall2024/data/World_Development_Indicators_(WDI).csv")

# Display the shape of the dataframe
print(df.shape)

df_filtered = df[['Country Code', 'Country', 'Region', 'Year', 'GDP per capita (current US$)']]

# Group by Region and Year, calculating the average GDP per capita
df_aggregated = df_filtered.groupby(['Region', 'Year']).agg(
    Average_GDP=('GDP per capita (current US$)', 'mean')
).reset_index()

# Round the Average GDP per capita to 2 decimal points
df_aggregated['Average_GDP'] = df_aggregated['Average_GDP'].round(2)
print(df_aggregated['Average_GDP'])

# Get the latest year data for each region
latest_year = df_aggregated['Year'].max()
df_latest = df_aggregated[df_aggregated['Year'] == latest_year]

# Sort the DataFrame by Average GDP in ascending order
df_latest = df_latest.sort_values(by='Average_GDP', ascending=True)
print(df_latest)

fig = px.bar(
    df_latest,
    x='Region',
    y='Average_GDP',
    color = 'Region',
    title=f'Average GDP per Capita by Region in {latest_year} (Sorted by GDP)',
    labels={'Average_GDP': 'Average GDP per Capita (Current US$)', 'Region': 'Region'},
    text='Average_GDP',
    height=600,
    width=1200
)

fig.show()
```

### QUESTION 3

Group the dataset by Region and Income Group, calculate the average, min, and max of GDP per Capita & life Expentancy, total, average, min, and max of population, then save the result to Excel.

```python
import pandas as pd

# Load the dataset
df = pd.read_csv("https://raw.githubusercontent.com/wcj365/python-stats-dataviz/refs/heads/master/fall2024/data/World_Development_Indicators_(WDI).csv")

# Display the shape of the dataframe
print(df.shape)

df_filtered = df[['Country Code', 'Country', 'Region', 'Income Group',
                   'Year', 'GDP per capita (current US$)',
                   'Life expectancy at birth, total (years)',
                   'Population, total']]

# Group by Region and Income Group and calculate the required statistics
grouped_stats = df_filtered.groupby(['Region', 'Income Group']).agg(
    Average_GDP=('GDP per capita (current US$)', 'mean'),
    Min_GDP=('GDP per capita (current US$)', 'min'),
    Max_GDP=('GDP per capita (current US$)', 'max'),
    Average_Life_Expectancy=('Life expectancy at birth, total (years)', 'mean'),
    Min_Life_Expectancy=('Life expectancy at birth, total (years)', 'min'),
    Max_Life_Expectancy=('Life expectancy at birth, total (years)', 'max'),
    Total_Population=('Population, total', 'sum'),
    Average_Population=('Population, total', 'mean'),
    Min_Population=('Population, total', 'min'),
    Max_Population=('Population, total', 'max')
).reset_index()

# Display the grouped statistics
print(grouped_stats)


# Save the result to an Excel file
grouped_stats.to_excel("region_income_group_stats.xlsx", index=False)
```
